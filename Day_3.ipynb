# Intro to Matplotlib

Matplotlib allows the creation of reproducible figure programmatically. 

* Matplotlib website - https://matplotlib.org/

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# iF using matplotlib on a local host to allow for plots to be shown in the notebook use the following command
# %matplotlib inline
```

```python
# Setting up the data

x = np.linspace(0,5,11)
y = x ** 2

x,y
```

## Simple plot

```python
plt.plot(x, y, "r");
plt.xlabel("X Label");
plt.ylabel("Y Label");
plt.title('Plot Title');

# semi colon above removes the output of the strings. makes it look cleaner
```

## Multiple plots on same canvas

```python
# plt.subplot(rows, columns, plot number)
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(y, x, "g--")
```

# Object Oriented Method

Just like everything else in programming object oriented is the way to go with it being more formal. This way is also makes the code look a lot cleaner and especially when dealing with multiple plots.

## Intro to OOM

The code looks more complicated but now there is full control over where to plot the axes and easily being able to scale and add more plots

```python
# Create fig
fig = plt.figure();

# Add axes
axes = fig.add_axes([0.2, 0.2, 0.8, 0.8]) # left, bottom, width, height

# Plot
axes.plot(x, y, "b")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title('Title')
```

## Inset Plot

```python
# Create fig
fig = plt.figure()

# Add axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Plot axes1
axes1.plot(x, y, "g")
axes1.set_xlabel("X Label 1")
axes1.set_ylabel("Y Label 1")
axes1.set_title('Title 1')

# Plot axes2
axes2.plot(y, x, "y")
axes2.set_xlabel('X Label 2')
axes2.set_ylabel("Y Label 2")
axes2.set_title('Title 2')

```

## Subplots

Subplots object is an automatic axis manager

```python
# Similiar to plt.figure except this way uses tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#plot
axes.plot(x, y, "r")
axes.set_xlabel("X Label")
axes.set_ylabel('Y Label')
axes.set_title("Plot Title")
```

## Specifying rows and columns with subplots object

```python
# Empty canvas of 1 by 2 (nrows , ncols)
fig, axes = plt.subplots(1, 2)
```

```python
# Iterating through array of axes

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
fig
```

A common issue is overlapping plots. Using `plt.tight_layout()` or `plt.tight_layout` which automatically adjusts the position

```python
# Create fig
fig, axes = plt.subplots(1, 2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

## Legend

Use the `label="label text"` when plotting or other objects are added to the figure, and then using the `legend` method without args to add the legend to the figure: 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
# Legend location options (4 examples)
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner


# Most common 
ax.legend(loc=0) # matplotlib decides the optimal location
fig
```

# Colors, Linewidths, Linetypes

```python
# Style line and color
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

## Colors with color= param

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

## Line and marking styles

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# Linestype options -, , -., :, steps and many more
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length and etc

# Possible marker symbols: marker = '+', 'o', '*', 's', ',', '.' and many more
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# Size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Plot range

Configure the ranges of the axes using the `set_ylim` and `set_xlim` methods or `axis('tight')` for automatically getting tightly fitted axes ranges

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Axes Ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('Tight')
axes[1].set_title("Tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Custom Axes Range");
```
